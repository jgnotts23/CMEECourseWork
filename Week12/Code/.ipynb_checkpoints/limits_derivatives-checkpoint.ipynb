## 2.4 Derivatives: the difference quotient
The difference quotient becomes the differential in the limit $h \to 0$ and describes the slope of a function $y = f(x)$ at a given point $x$:
\begin{equation}
y'(x) = \frac{dy}{dx} = \underset{h \to 0}{\text{lim}}\frac{y(x + h) - y(x)}{h}
\end{equation}

```python
# An example of a map and its derivative
# The slope of a curve is found from its derivative
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import math

x = np.linspace(-4, 4, 100)

def quad (x):
    return x**2

plt.plot(x, quad(x), label="y = x^2")
plt.plot(x, 2*x, label="y' = 2x")
plt.legend()
```

**Notation:** the limit value of the difference quotient is called the derivative of a function $f(x)$.
Derivatives are denoted by:
\begin{equation}
y(x) , \ \frac{dy}{dx} , \ \frac{df}{dx} , \ \frac{d}{dx}f(x)
\end{equation}

```python
%matplotlib notebook

t = np.linspace(0, 10, 100)

plt.plot()
axes = plt.gca()
axes.set_xlim([0,10])
axes.set_ylim([0,500])

for i in range(100):
    plt.plot(t, 2**t, 'k')
    plt.plot(t, 3**t, 'g')
    plt.plot(t, math.e**t, 'b')
    plt.plot(t, 10**t, 'r')
```
